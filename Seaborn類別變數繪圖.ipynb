## <span style ='color:green'> __Categorical distribution plots:__ </span>
> # boxplot()

```python
import seaborn as sns
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
# ax = sns.boxplot(x="day", y="total_bill", data=tips)
ax = sns.boxplot(x="day", y="total_bill", hue="smoker",\
                  data=tips, palette="Set3")

g = sns.catplot(x="sex", y="total_bill",\
                 hue="smoker", col="time",\
                 data=tips, kind="box",\
                 height=4, aspect=.7);
```

```python
# Barplot
import seaborn as sns
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
# ax = sns.barplot(x="day", y="total_bill", data=tips)
# ax = sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
# ax = sns.barplot(x="time", y="tip", data=tips,order=["Dinner", "Lunch"]) # control bar order
ax = sns.barplot("size", y="total_bill", data=tips,palette="Blues_d")

g = sns.catplot(x="sex", y="total_bill",\
                 hue="smoker", col="time",\
                 data=tips, kind="bar",\
                 height=4, aspect=.7);
```

```python
# countplot
import seaborn as sns
sns.set(style="darkgrid")
titanic = sns.load_dataset("titanic")
# ax = sns.countplot(x="class", data=titanic)
ax = sns.countplot(x="class", hue="who", data=titanic, palette="Set1")

g = sns.catplot(x="class", hue="who", col="survived",\
                 data=titanic, kind="count",\
                 height=4, aspect=.7);
```
